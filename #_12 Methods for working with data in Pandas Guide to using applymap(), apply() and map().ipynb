```python
# task 3
import pandas as pd

count_df = pd.read_csv(r'C:\Users\dell\Desktop\img\money.csv')

print(count_df.salary)

count_df.salary = count_df.salary.map(lambda x: x + x*10/100)

def full_name(row):
    return row['first_name'] + ' ' + row['last_name']

count_df['full_name'] = count_df.apply(full_name, axis = 1)

def set_seniority(x):
    if x.salary < 75000:
        x.seniority = 'junior'
        return x.seniority
    elif x.salary > 75000 and x.salary < 130000:
        return 'middle'
    else:
        return 'senior'

count_df["seniority"] = count_df.apply(set_seniority, axis = 'columns') 


count_df.loc[:, ['full_name', 'salary', 'seniority']]
```

```python
import pandas as pd

count_df = pd.read_csv(r'C:\Users\dell\Desktop\img\money.csv')

print(count_df.salary)

count_df.salary = count_df.salary.map(lambda x: x + x*10/100)

def full_name(row):
    return row['first_name'] + ' ' + row['last_name']

count_df['full_name'] = count_df.apply(full_name, axis = 1)

def set_seniority(x):
    if x < 75000:
        x = 'junior'
        return x
    elif x > 75000 and x < 130000:
        return 'middle'
    else:
        return 'senior'

count_df["seniority"] = count_df['salary'].apply(set_seniority) 


count_df.loc[:, ['full_name', 'salary', 'seniority']]
```
