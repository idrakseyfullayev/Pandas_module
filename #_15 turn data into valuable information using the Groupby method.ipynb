```python
import pandas as pd

count_df = pd.read_csv(r'C:\Users\dell\Desktop\img\money.csv')

def emp_id(x):
    if x.age < 30:
        x.employee_id = 1
    elif x.age > 30 and x.age < 45:
        x.employee_id = 2
    else:
        x.employee_id = 3
     
    return x

count_df = count_df.apply(emp_id, axis=1)

count_df
```
